# Visualize the distribution of the data
- Identical metrics can represnt very different data distributions


## Seaborn
- **seaborn.distplot()** : Swiss Army knife when visualizing distribution of the variable. use it to draw **histogram** or **kernel density estimation (KDE)** 
- use displot to see how its integrates histogram with KDE, which is a smooth version of a histogram


```python
# 10 year treasury : Trend & distribution
import pandas as pd
from pandas_datareader.data import DataReader
from datetime import date
import matplotlib.pyplot as plt
```

```python

ty10 = DataReader('DGS10', 'fred', date(1962,1,1))
ty10.info()
```

- The number of indices do not match the number of values. To correct this we need to eliminate the missing values
- .dropna() : drops all rows with atleast one missing value
- .fillna() : replaces all missing values with the new values


```python
ty10.describe()
```

- descriptive stats shows that mean is bit higher then the median, suggesting that the data has a positive skew with the longer tail at the right

```python
ty10.dropna(inplace=True)
```

```python
ty10.plot(title='10-year Treasury')
plt.tight_layout()
```

## Seaborn distribution plot
- In contrast to the timeseries perspective we can draw seaborn distribution plot to draw histogram together with the Kernel Density Estimate
- Histogram creates bins or ranges & plots the counts of the subset of the data that fall into each bin to reflect the frequency distribution
- **Kernel Density Estimate** provides a smooth approximation of the frequency distribution. It is particularly used to **compare different distribution** 

```python
import seaborn as sns
sns.distplot(ty10)
```

```python
## add vertical lines that marks important values(in this case median)
ax = sns.distplot(ty10)
ax.axvline(ty10['DGS10'].median(), color='black', ls='--')
```

```python
# Load the file into growth
growth = pd.read_csv('data/income_growth.csv', parse_dates=['DATE']).set_index("DATE")

# Inspect the summary statistics for the growth rates
growth.describe()

# Iterate over the three columns
for column in growth.columns:
    sns.distplot(growth[column], hist=False, label=column)
    
# Show the plot
plt.show()
```

```python
# Create inc_per_capita
income = pd.read_csv('data/per_capita_income.csv')
inc_per_capita = income['Income per Capita']

# Filter out incomes above the 95th percentile
inc_per_capita = inc_per_capita[inc_per_capita < inc_per_capita.quantile(.95)]

# Plot histogram and assign to ax
ax = sns.distplot(inc_per_capita)

# Highlight mean
ax.axvline(inc_per_capita.mean(), color='b')

# Highlight median
ax.axvline(inc_per_capita.median(), color='g')

# Show the plot
plt.show()
```
